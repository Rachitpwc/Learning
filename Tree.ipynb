TREE

Nonlinear data structure with hierarchial relationships between its elements without having any cycle, it is basically reversed from a real life tree.

```python
# Create a Tree
class Tree:
    def __init__(self, data, children = []):
        self.data = data 
        self.children = children
    
    def __str__(self, level=0):
        ret = "  " * level + str(self.data) + "\n"
        for child in self.children:
            ret += child.__str__(level+1)
        return ret
    
    def addChild(self, TreeNode):
        self.children.append(TreeNode)


trep1 = Tree('Drinks',[])
cold = Tree('Cold',[])
hot = Tree('Hot',[])
trep1.addChild(cold)
trep1.addChild(hot)
print(trep1)
tea = Tree('Tea',[])
coffee = Tree('Coffee', [])
cola = Tree('Cola', [])
fanta = Tree('Fanta',[])
cold.addChild(cola)
cold.addChild(fanta)
hot.addChild(tea)
hot.addChild(coffee)
print(trep1)
```
